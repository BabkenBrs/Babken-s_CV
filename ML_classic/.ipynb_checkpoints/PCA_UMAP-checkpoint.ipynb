```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set(palette='Set2')
from sklearn.decomposition import PCA

import scipy as sp
import mnist_reader
```

```python
from umap.umap_ import UMAP
from umap import umap_
from sklearn.model_selection import train_test_split
```

---

Fashion-MNIST — набор данных изображений одежды, обуви и аксессуаров, состоящий из обучающего набора из 60 000 изображений и тестового набора из 10 000 изображений. Каждое изображение имеет размер 28x28 пикселей в оттенках серого.

```python
X_train, y_train = mnist_reader.load_mnist(
    'fashion-mnist/data/fashion', kind='train'
)
X_test, y_test = mnist_reader.load_mnist(
    'fashion-mnist/data/fashion', kind='t10k'
)
```

Объединим все данные и зададим названия классов

```python
data = np.array(np.vstack([X_train, X_test]), dtype=np.float64) / 255.0
target = np.hstack([y_train, y_test])
classes = [
    'Футболки/топы',
    'Брюки',
    'Свитеры',
    'Платья',
    'Пальто',
    'Сандалии',
    'Рубашки',
    'Кроссовки',
    'Сумки',
    'Сапоги'
]
```

Посмотрите на размер данных и напечатайте несколько изображений

```python
data.shape
```

```python
plt.figure(figsize=(10, 4))
for i in range(40):
    plt.subplot(4, 10, i + 1)
    plt.imshow(data[i].reshape((28, 28)), cmap='gray')
    plt.axis('off')
```

UMAP для сжатия в двумерное пространство.

Визуализируем полученные результаты. В качестве цвета точки используем ее класс.

```python
model = UMAP(n_components=2).fit(X_train)
X_umap_train = model.transform(X_train)
X_umap_test = model.transform(X_test)
```

```python
plt.figure(figsize=(15, 15))
plt.scatter(X_umap_train[:, 0], X_umap_train[:, 1], label='train', c = y_train, cmap = 'tab10')
plt.scatter(X_umap_test[:, 0], X_umap_test[:, 1], s=120, label='test', marker='^', 
            edgecolors='black', c = y_test, cmap = 'tab10')

plt.legend()


cb = plt.colorbar()
cb.set_ticks(np.arange(len(classes)))
cb.set_ticklabels(classes)

plt.xlabel('Латентная координата 1')
plt.ylabel('Латентная координата 2')

plt.title('UMAP train и test без уителя', fontsize = 20, fontweight = 'bold')
```

Проинтерпретируйте полученные результаты. Удалось ли выделить кластеры? Что можно сказать про взаиморасположение кластеров в двумерном пространстве?

По графику видно, что наши данные разделились на 4 остравка, каждый из которых имеет определенный смысл. Так например сандали, кросовки и сапоги объединились в один островок и образовали кластер обувь. Аналогично сумки образовали отдельный кластер.

Обучите UMAP, передав ему также метки классов, и визуализируйте результаты. Как изменились результаты?

```python
model = UMAP(n_components=2).fit(X_train, y_train)
X_umap_train = model.transform(X_train)
X_umap_test = model.transform(X_test)
```

```python
plt.figure(figsize=(15, 15))
colors = ['red', 'black', 'green', 'blue', 'orange', 'brown', 'yellow', 'violet', 'gray', 'white']
plt.scatter(X_umap_train[:, 0], X_umap_train[:, 1], label='train', cmap='tab10', c = y_train)
plt.scatter(X_umap_test[:, 0], X_umap_test[:, 1], s=120, label='test', marker='^', 
            edgecolors='black', cmap='tab10', c = y_test)

plt.legend()


cb = plt.colorbar()
cb.set_ticks(np.arange(len(classes)))
cb.set_ticklabels(classes)

plt.xlabel('Латентная координата 1')
plt.ylabel('Латентная координата 2')

plt.title('UMAP train и test с учителем', fontsize = 20, fontweight = 'bold')
```

Передав метки в UMAP мы видим, что разделение происходит ровно на 10 кластеров. 

**Вывод:**


1) UMAP хорошо справляется с задачей, как в обучении с учителем, так и без, что видно по результатам, представленным на графиках.

2) Что примечательно, без учителя UMAP разбивает всё на категории (верхняя одежда, обувь и тд), а с учителем на более узкие классы (кроссовки, сапоги, свитер, брюки и тд).

Рассмотрим датасет **Leaf Classification**:
	
https://www.kaggle.com/c/leaf-classification
	
Данные содержат 1584 изображений образцов листьев (16 изображений для 99 видов). По ссылке доступно подробное описание данных. Для вашего удобства размер	некоторых изображении был изменен, в результате чего все изображения имеют одинаковый размер $170 \times 250$. 

```python
plt.figure(figsize = (10, 4))
for i in range(1, 5):
    plt.subplot(2, 2, i)
    plt.imshow(plt.imread('Leaf_Images/scaled_images/'+ str(i) + '.jpg'))
    plt.axis('off')
```

```python
y_train = pd.read_csv('Leaf_Images/train_labels.csv')
y_train.head()
```

```python
X = []
for i in range(1, 1585):
    X.append(plt.imread('Leaf_Images/scaled_images/'+ str(i) + '.jpg'))
X = np.array(X)
```

```python
X
```

```python
X_train, X_test = X[y_train.id - 1], X[np.array(list(set(range(1, 1585)) - set(y_train.id))) - 1]
```

```python
X_train.shape
```

```python
pca = PCA(n_components=30)
transform_X_train = pca.fit_transform(X_train.reshape(990, -1))
```

Доля объясненной дисперсии каждой компонентой

```python
pca.explained_variance_ratio_
```

Кумулятивная доля объясненной дисперсии

```python
pca.explained_variance_ratio_.cumsum()
```

```python
print(f'Доля объясненной дисперсии 30 компонент = {sum(pca.explained_variance_ratio_)}')
```

```python
plt.figure(figsize=(10, 10))
plt.suptitle('Визуализация 30 главных компонент', fontsize = 15, fontweight = 'bold')

for i in range(0, 30):
    plt.subplot(5, 6, i + 1)
    plt.imshow(pca.components_[i, :].reshape(170, 250), interpolation='none', cmap='gray')
    plt.axis('off')
```

Как мы знаем, главные компоненты стараются содержать в себе информацию об объектах классификации. К сожалению, их всего 30, именно поэтому доля объясненной дисперсии ниже 1. Однако, даже 30 главных компонент способны дать нам определенную информацию об объектах. Когда мы будем проецировать объекты на эти компоненты, будет видно (конечно, с определенной ошибкой), к какому классу принадлежит объект. Иначе говоря, с какой из вышепоказанных картин он схож. Главные компоненты имеют разные формы, разную насыщенность в тех или иных местах, разную площадь.

```python
sns.set(font_scale=1.3)
plt.figure(figsize=(13, 10))
plt.scatter(transform_X_train[:, 0], transform_X_train[:, 1], c = pd.factorize(y_train.species)[0], alpha=0.7, cmap='Set1')
plt.xlabel('Проекция на первую главную компоненту')
plt.ylabel('Проекция на вторую главную компоненту')

plt.title('Проекции на первые две компоненты', fontsize = 20, fontweight = 'bold')
plt.axis('equal')
plt.show()
```

Закономерности сложно обнаружить. Но это можно объяснить тем, что у нас достаточно низкая доля объясненной дисперсии

UMAP на основе исходных признаков

```python
model = UMAP(n_components=2).fit(X_train.reshape(990, -1))
X_umap_train = model.transform(X_train.reshape(990, -1))
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(X_umap_train[:, 0], X_umap_train[:, 1], label='train', cmap='tab10', c = pd.factorize(y_train.species)[0], alpha = 0.6)

plt.xlabel('Проекция на первую главную компоненту')
plt.ylabel('Проекция на вторую главную компоненту')

plt.title('UMAP на основе исходных признаков', fontsize = 20, fontweight = 'bold')
```

```python
model_transform = UMAP(n_components=2).fit(transform_X_train)
transform_X_umap_train = model_transform.transform(transform_X_train)
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(transform_X_umap_train[:, 0], transform_X_umap_train[:, 1], label='train', cmap='tab10', c = pd.factorize(y_train.species)[0], alpha = 0.6)

plt.xlabel('Проекция на первую главную компоненту')
plt.ylabel('Проекция на вторую главную компоненту')

plt.title('UMAP по проекциям на первые 30 компонент', fontsize = 20, fontweight = 'bold')
```

**Вывод:**

1. По поводу первого графика. Мы визуализировали обучающую часть данных в проекции на две первых главных компоненты. Видим, что картина получилось не очень хорошим, но это объясняется тем, что у нас достаточно низкая доля объясненной дисперсии.

2. По поводу работы UMAP. Графики получились ужасными: не произошла нормально кластеризация, не понятен смысл разделения. Трудно понять, почему так произошло, но факт остается фактом. Может быть это из-за того, что у нас огромная разновидность данных.
