# Примеры работы классификатора

```python
classificator = OpenEyesClassificator("/content/optim_model_weights.pt")
```

```python
def ShowImageScore(img_path, classificator, is_open = True):
  img = Image.open(img_path)
  plt.imshow(img, cmap = 'gray')
  print(f"Вероятность того, что глаз открыт: {classificator.predict(img_path)}")
  if is_open:
    print("На изображении открытый глаз")
  else:
    print("На изображении закрытый глаз")
```
